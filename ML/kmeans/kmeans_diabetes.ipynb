```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('diabetes.csv')
print(data.head(3))
```

```python
x=data[['Glucose','BloodPressure','BMI','Age']]
y=data['Outcome']
```

```python
x=x.values
y=y.values
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='jet')
plt.show()
```

```python
model=KMeans(n_clusters=2)
```

```python
model.fit(x)
```

```python
lb=model.labels_
lb1=set(lb)
labels=list(lb1)
print(labels)
print(y)
```

```python
centroids=model.cluster_centers_
print(centroids)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='jet',label='vals')
plt.scatter(centroids[:,0],centroids[:,1],marker='*',color='black',label='centroids')
plt.title("clustered")
plt.legend()
plt.show()
```
