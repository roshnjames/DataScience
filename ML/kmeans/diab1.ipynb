```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import cluster
```

```python
data=pd.read_csv('diabetes.csv')
print(data.head(3))
```

```python
model=cluster.KMeans(n_clusters=2)
```

```python
x=data[['Glucose','BloodPressure','BMI','Age']]
y=data['Outcome']
```

```python
model.fit(x)
```

```python
x=x.values
y=y.values
```

```python
lb1=model.labels_
lb2=set(lb1)
labels=list(lb2)
print(labels)
```

```python
c=model.cluster_centers_
print(c)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='viridis')
plt.scatter(c[:,0],c[:,1],marker='+',color='black')
plt.show()
```
