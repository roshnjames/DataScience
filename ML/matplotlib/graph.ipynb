```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[1,2,6,18]
y=[3,10,12,20]
```

```python
plt.plot(x,y,ls='--',marker='o',markerfacecolor='green', color='red', markersize=10)
plt.title('line graph')
plt.xlabel('xvalues')
plt.ylabel('yvalues')
plt.grid(True)

plt.show()
```

```python
x1=[12,14,16,18,20,22,24]
y1=[100,200,250,400,300,450,500]
```

```python
plt.plot(x1,y1,ls='--',color='blue',marker='o',markersize=11,markerfacecolor='red',)
plt.title('box values')
plt.xlabel('xxx')
plt.ylabel('yyy')
plt.grid()
plt.legend('blue')
plt.show()
```

```python
with open('vals.txt') as vals:
    cords=vals.read().splitlines()

xvals=[]
yvals=[]

for d in cords:
    x,y=map(int,d.split())
    xvals.append(x)
    yvals.append(y)

print(xvals)
print(yvals)
```

```python
plt.plot(xvals,yvals,ls='--',color='red',marker='o',markersize='11',markerfacecolor='blue')
plt.title("qqq")
plt.xlabel('xxx')
plt.ylabel('yyy')
plt.grid(1)
plt.show()
```

```python
yvals2=[11,22,15,34,22,40]

plt.plot(xvals,yvals,marker='o',ls='--',color='red',label='reg')
plt.plot(xvals,yvals2,color='blue',marker='^',label='sec')
plt.title("2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(xvals,yvals)
plt.title("1")

plt.subplot(2,2,2)
plt.plot(xvals,yvals2)
plt.title("2")

plt.tight_layout()
plt.show()
```

```python
langs=['java','python','php','js','c#','c++']
popularity=[22.2,17.6,8.8,8,77,6.7]
colors=['yellow','blue','red','cyan','orange','yellow']
```

```python
plt.barh(langs,popularity,color=colors)
plt.xlabel("x")
plt.grid(1)
plt.show()
```

```python
men=(22,30,35,35,26)
women=(25,32,30,35,29)
x=np.arange(len(men))
width=0.35
```

```python
plt.bar(x,men,width=width,color='red',label='men')
plt.bar(x+width,women,width=width,color='green',label='women')
plt.title("men vs women")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.pie(popularity,labels=langs,autopct='%1.1f%%',startangle=90,)
plt.axis("equal")
plt.show()
```
